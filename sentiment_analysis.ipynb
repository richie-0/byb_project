```python
import numpy
import spacy
import pandas as pd
```

```python
df = pd.read_csv("amazon_product_reviews.csv")
df.head()
```

Cleaning

```python
# We only need select columns for the sentiment analysis.
cleaned = df[['reviews.text', 'reviews.title']]
cleaned.head()
```

```python
# Check for any null values.
cleaned.isnull().sum()
```

```python
nlp = spacy.load('en_core_web_md')
```

Preprocess

```python
# Lematize text and filter out punct and stop words.
cleaned['processed.text'] = cleaned['reviews.text'].apply(lambda sample: ' '.join([token.lemma_.lower() for token in nlp(sample.strip()) if not token.is_punct and not token.is_stop]))
cleaned.head()
```

Polarity

```python
from collections import defaultdict
from textblob   import TextBlob
from wordcloud import WordCloud
import matplotlib as plt
%matplotlib inline
```

```python
# Polarity for preprocessed data.

pos_words = defaultdict(int)
neg_words = defaultdict(int)

def analyze_polarity(sample):
    
    blob = TextBlob(sample)

    for word in blob.words:
        polarity = word.sentiment.polarity
        
        if polarity > 0:
            pos_words[word] =+ 1

        elif polarity < 0:
            neg_words[word] =+ 1
```
